```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv('CarPrice_Assignment.csv')
df
```

```python
df.info()
```

```python
df['doornumber'].unique()
```

```python
df['doornumber']= df['doornumber'].map({
    'two':2,
    'four':4
})
```

```python
df['doornumber'].astype
```

```python
df['drivewheel'].unique()
```

```python
df['drivewheel']=df['drivewheel'].map({
    'rwd':2,
    'fwd':1,
    '4wd':3

})
```

```python
df['enginelocation'].unique()
```

```python
df['enginelocation']=df['enginelocation'].map({
    'front':1,
    'rear':2
})
```

```python
df['cylindernumber'].unique()
```

```python
df['cylindernumber']=df['cylindernumber'].map({
    'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8
})
```

```python
df.info()
```

```python
int_flot=df.select_dtypes(include=['int64','Float64']).columns
print(int_flot)
```

```python
x=df[[ 'doornumber', 'drivewheel', 'enginelocation',
       'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight',
       'cylindernumber', 'enginesize', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']]
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
y_pred=regression.predict(x_test)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mean absolute error",mae)
print("r2 score",score)
plt.scatter(y_test,y_pred)
```

```python
df["y_test_pred"]= pd.DataFrame(regression.predict(scaler.transform(x)))
df
```
